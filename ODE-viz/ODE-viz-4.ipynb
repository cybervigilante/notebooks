```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python

def make_one_plot(fn=1, xmin=-1, xmax=1, ymin=-1, ymax=1, xtick_mult=1, ytick_mult=1, figname="plot", usecolor=False, clist=None):
    global counter
    savename = figname + str(counter) + ".pdf"
    counter = counter + 1
    FIGSIZEX = 3
    FIGSIZEY = 3

    xs = np.linspace(xmin, xmax, num=200)
    if not clist:
        clist = [-1 + x*1./3 for x in range(7)] 
    for c in clist:
        f = lambda x: fn(x, c)
        ys = np.array(list(map(f, xs)))
        if not usecolor:
            plt.plot(xs, ys, color='k')
        if usecolor:
            plt.plot(xs, ys)
    
    axes = plt.gca()
    axes.spines['top'].set_visible(False)
    axes.spines['right'].set_visible(False)
    #axes.spines['left'].set_position('zero')
    #axes.spines['bottom'].set_position('zero')
    #xticks = [-10 + .5*x for x in range(41)]
    xticks = [xtick_mult*(-10 + x) for x in range(21)]
    yticks = [ytick_mult*(-20 + x) for x in range(41)]
    plt.xticks(xticks)
    plt.yticks(yticks)
    plt.xlim([xmin, xmax])
    plt.ylim([ymin, ymax])
    plt.xlabel("x")
    plt.ylabel("y")
    axes.grid(True, linestyle='--')
    
    fig = plt.gcf()
    fig.set_size_inches(FIGSIZEX, FIGSIZEY)
    fig.savefig(savename, bbox_inches='tight')
```

```python
counter = 1
def fn(x,c):
    return math.cos(5*x) + 0.4*x* math.sin(5*x)
make_one_plot(fn, xmin=0, xmax=10, ymin=-4, ymax=4, figname="ex47_10", clist=[0])
```

```python
counter = 1
def fn(x,c):
    if c == 0:
        return -math.cos(2*x) / 5
    if c == 1:
        return 8*math.exp(-x)/5 - 2*math.exp(-4*x)/5
    if c == 2:
        return fn(x,0) + fn(x,1)
make_one_plot(fn, xmin=0, xmax=10, ymin=-1, ymax=1.5, ytick_mult=0.5, figname="ex47_10", clist=[0,1,2])
```
